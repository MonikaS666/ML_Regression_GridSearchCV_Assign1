```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
#n_estimators-->int, default=50
#learning_rate: float, default=1.0
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[10,50,80,100],"learning_rate":[0.3,0.1,1.0,1.5,2]}
grid=GridSearchCV(AdaBoostRegressor(),param_grid,refit=True,n_jobs=-1,verbose=3)
grid.fit(Xtrain,Ytrain)
```

```python
print(grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(grid.cv_results_)
table
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
#␣change the paramter,play with it.
print("Future_Prediction={}".format(Future_Prediction))
```
